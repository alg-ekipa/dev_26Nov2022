```python
import pandas as pd
```

```python
listings_df= pd.read_csv('airbnb_new_york_city_listings.csv')
listings_df
```

```python
listings_df.dtypes
```

```python
#listings_df['last_review'] =  listings_df['last_review'].astype('datetime64')
listings_df['last_review']=pd.to_datetime(listings_df['last_review'])
listings_df.dtypes
```

```python
listings_df['price']=listings_df['price'].astype('float')
listings_df.dtypes
```

```python
listings_df['last_review_year']=listings_df['last_review'].dt.year
#kreiramo novi stupac, uzimamo godinu iz stupca last review

listings_df
```

```python
listings_df.isnull().sum()
```

```python
listings_df['name'].fillna('No name listing', inplace=True)   #.fillna naredba za dodavanje vrijednosti praznim poljima
```

```python
listings_df.isnull().sum()
```

```python
listings_df.dtypes
```

```python
listings_df['host_name'].fillna('J.Doe', inplace=True)
listings_df['last_review'].fillna('01-01-1970', inplace=True)
listings_df['reviews_per_month'].fillna(0.0, inplace=True)
listings_df['last_review_year'].fillna(0.0, inplace=True)
```

```python
listings_df.isnull().sum()
```

```python
listings_df.dtypes
```

```python
listings_df['last_review_year']=listings_df['last_review_year'].astype('int64')
```

```python
listings_df.dtypes
```

```python
listings_df['name']=listings_df['name'].str.strip()   #čišćenje zankova prije i poslije (npr \n)
listings_df['host_name']=listings_df['host_name'].str.strip()
listings_df['neighbourhood_group']=listings_df['neighbourhood_group'].str.strip()
listings_df['neighbourhood']=listings_df['neighbourhood'].str.strip()
listings_df['room_type']=listings_df['room_type'].str.strip()
```

```python
listings_df.dtypes
```

```python
listings_df.isnull().sum()
```
