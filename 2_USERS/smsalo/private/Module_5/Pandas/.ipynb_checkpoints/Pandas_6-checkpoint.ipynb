## mijenjanje  naziva stupaca

```python
ocjene_df.columns = ['Ime', 'Prezime', 'Spol', 'Dob', 'Sati_vjezbanja', 'Sati_ucenja', 'Ocjena', 'Adresa']
ocjene_df.columns
```

### metoda MAP

```python
ocjene_df.columns
```

### .lower

```python
ocjene_df.columns=map(str.lower, ocjene_df.columns)
ocjene_df.columns
```

### .upper()

```python
ocjene_df.columns = [x.upper() for x in ocjene_df.columns]
ocjene_df.columns
```

### .title()

```python
ocjene_df.columns = [x.title() for x in ocjene_df.columns]
ocjene_df.columns
```

# FILTRIRANJE po uvjetima

```python
(ocjene_df['Spol']=='female') & (ocjene_df['Ocjena']>70)
```

## pozicije traženih uvjeta

```python
ocjene_df.loc[(ocjene_df['Spol']=='female') & (ocjene_df['Ocjena']>70)]
```

### sortiranje
.sort_values(by=...)

```python
ocjene_df.sort_values(by='Sati_Vjezbanja', ascending=True)
```
