```python
import numpy as np
```

```python
x1 = np.linspace(0,9,10)
x1
```

```python
y1=2*x1
y1
```

```python
plt.plot(x1,y1)
```

```python
plt.plot(x1,y1)
plt.title('Linearna funkcija')
plt.xlabel('x')
plt.ylabel('y')
```

```python
x2=np.linspace(0,10,20)
y2=x2*x2
plt.plot(x2,y2,'red')
plt.title('Kvadratna funkcija')
plt.xlabel('x')
plt.ylabel('y')
```

```python
plt.subplot(1,2,1)   #argumenti: (broj redaka, broj stupaca, pozicija)
plt.plot(x1,y1) #Linerna funkcija na poziciji 1

plt.subplot(1,2,2)   #argumenti: (broj redaka, broj stupaca, pozicija)
plt.plot(x2,y2,'r')
```

```python
plt.subplot(2,2,1)   #argumenti: (broj redaka, broj stupaca, pozicija)
plt.plot(x1,y1) #Linerna funkcija na poziciji 1

plt.subplot(2,1,2)   #argumenti: (broj redaka, broj stupaca, pozicija)
plt.plot(x2,y2,'r')
```

```python
casedistribution_df = pd.read_csv('casedistribution.csv')
casedistribution_df
```

```python
casedistribution_df.plot(
    kind = 'line',
    x='dateRep',
    y='cases',
    figsize=(10,5),
    style='-b',
    legend=True,
    title='Broj slučajeva'
)
```

```python
casedistribution_df.groupby('dateRep', as_index=False).mean().plot(
    kind = 'line',
    x='dateRep',
    y='cases',
    figsize=(10,5),
    style='-b',
    legend=True,
    title='Broj slučajeva')
```

```python
casedistribution_df.groupby('dateRep', as_index=False).mean() 
```

```python
casedistribution_df.plot(
    kind = 'line',
    x='dateRep',
    y=['cases','deaths'],
    xlabel='Datum',
    ylabel='Broj zaraženih',
    
    figsize=(10,5),
    #style='-b',
    color=['green','red'],
    linestyle='solid',
    legend=True,
    title='Broj slučajeva'
)
```

```python
casedistribution_df.groupby('dateRep', as_index=False).mean().plot(
    kind = 'line',
    x='dateRep',
    y=['cases','deaths'],
    figsize=(10,5),
    subplots='True',
    layout(1,2),
    color=['green','red'],
    linestyle='solid',
    title='Broj slučajeva'
)
```

```python
novi_slucajevi_prosjek=casedistribution_df.pivot_table(
    index=casedistribution_df.index,
    columns='countriesAndTerritories',
    values='cases').rolling(7).mean()

okvir, axes = plt.subplots(1,3, figsize=(15,5))

novi_slucajevi_prosjek[['Germany']].plot(
    ax=axes[0],
    style='-r'
)

novi_slucajevi_prosjek[['France']].plot(
    ax=axes[1],
    style='.b'
)

novi_slucajevi_prosjek[['Croatia']].plot(
    ax=axes[2],
    style='--'
)
```

```python
data={
    'apple':10,
    'orange':15,
    'lemon':5,
    'lime':20
}
x=list(data.keys())
x
```

```python
y=list(data.values())
y
```

```python
figure, axes = plt.subplots(1,4, figsize=(9,3))
axes[0].plot(x,y)
axes[1].bar(x,y)
axes[2].scatter(x,y)
axes[3].pie(y)
```
